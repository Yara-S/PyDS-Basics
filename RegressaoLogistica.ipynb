```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_train = pd.read_csv("titanic_train.csv")
```

```python
#objetivo: criar um modelo de regressao logistica para tentar prever que passageiros sobreviveriam ao titanic

titanic_train.head()
```

```python
titanic_train.isnull().head()
```

```python
#nos livrando das linhas com dados nulos
titanic_train.dropna(inplace=True)
```

```python
#tratando o sexo para trata-lo como numero
sex = pd.get_dummies(titanic_train["Sex"], drop_first=True) #dropa o feminino, infere-se que tudo que nao for M é FEM
```

```python
#retirar as colunas que nao vao agregar na regressao

titanic_train.drop(["PassengerId", "Name", "Sex", "Ticket", "Cabin", "Embarked"], axis=1, inplace=True)
```

```python
titanic_train.head()
```

```python
#concatenar com o sexo

titanic_train = pd.concat([titanic_train, sex], axis=1)
titanic_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(titanic_train.drop(["Survived"],axis=1), titanic_train["Survived"], test_size=0.3)
```

```python
log = LogisticRegression()
log.fit(x_train, y_train)

predict = log.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predict)) #faz um report baseado na precisao do predict
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predict))
```
