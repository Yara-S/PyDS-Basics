# Plots de distribuição

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
graf1 = sns.load_dataset("tips")
```

```python
graf1.head()
```

```python
sns.distplot(graf1["total_bill"])
```

```python
sns.distplot(graf1["total_bill"], kde=False)
```

```python
sns.jointplot("total_bill", "tip", data=graf1)
```

```python
sns.jointplot("total_bill", "tip", data=graf1, kind="hex")
```

```python
sns.pairplot(graf1) #usa todas as variaveis numericas
```

```python
sns.pairplot(graf1, hue="sex")
```

```python
sns.rugplot(graf1["total_bill"])
```

# Plots Categoricos

```python
sns.barplot(x="sex",y="tip", data=graf1)#a linha preta é o desvio padrao
```

```python
sns.countplot(x="sex", data=graf1)
```

```python
sns.boxplot(x="day",y="total_bill", data=graf1)
```

```python
sns.violinplot(x="day",y="tip", data=graf1)
```

```python
sns.stripplot(x="day",y="tip", data=graf1, hue="sex", dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=graf1)
```

```python
sns.catplot(x="sex", y="total_bill", data=graf1, kind="bar") #muda conforme o kind
```

# Plots Matriciais

```python
graf2 = sns.load_dataset("flights")
graf2.head()
```

```python
aux = graf2.corr()
graf2.corr()
```

```python
sns.heatmap(aux)
```

```python
graf21 = graf2.pivot_table(values="passengers", index="month", columns="year")
```

```python
graf21
```

```python
sns.heatmap(graf21, cmap="Blues")
```

# Plots de regressão

```python
sns.lmplot(x="total_bill", y="tip", data=graf1, hue="sex", markers=["o", "v"])
```
