```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline 
#imprimir o grafico na propria linha
```

```python
x = np.linspace(0,100,10)
y = x ** 2
```

```python
plt.plot(x, y)
plt.show() #tira a saida antes do grafico
```

```python
y = x + x
plt.plot(x,y)
```

```python
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Titulo do Grafico")
plt.plot(x,y, color="r")
```

```python
plt.plot(x,y, 'g--')
```

```python
plt.subplots(nrows=1, ncols=2)
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y, color="r")
plt.subplot(1,2,2)
plt.plot(x,y, 'g--')
```

# Plotando duas funções no mesmo grafico

```python
figure = plt.figure()
axes1 = figure.add_axes([0.1,0.1,0.8,0.8])  #percentual de posicionamento [esquerda, baixo, largura, altura]


y = x * x
axes1.plot(x,y)

axes2 = figure.add_axes([0.2, 0.5, 0.3, 0.3])
axes2.plot(y,x, "r")
```

```python
figure, ax = plt.subplots()
ax.plot(x, x*2, 'b--', label="Dados1")
ax.plot(x, x*2+x, 'r-.', label="Dados2")
ax.set_xlabel("Eixo X")
ax.set_title('Titulo')
ax.legend(loc=4)
```

```python
figure, axe = plt.subplots(1,2)
axe[0].plot(x,y, 'g')
```

```python
figure, axe1 = plt.subplots(figsize=(12,3))
axe1.plot(y,x)
axe1.set_xlim([0,10000])
axe1.set_ylim([0, 100])
```

```python
figure.savefig("exemplo-funcao.png")
```
