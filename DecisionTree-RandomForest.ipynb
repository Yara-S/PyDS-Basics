```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("kyphosis.csv")
```

```python
data.head()
```

# Procurar a variavel que divide melhor

```python
sns.pairplot(data, hue="Kyphosis")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.drop(["Kyphosis"], axis=1), data["Kyphosis"],test_size=0.3 )
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
predict = dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predict))
```

# Floresta Aleatoria

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data2 = RandomForestClassifier(n_estimators=200)
```

```python
data2.fit(x_train, y_train)
```

```python
predict2 = data2.predict(x_test)
```

```python
#como o dataset em uestao é pequeno, as diferencas sao pequenas
#na partica, arvores aleatorias sao mais precisas
print(classification_report(y_test, predict2))
```
